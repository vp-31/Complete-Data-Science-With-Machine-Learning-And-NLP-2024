## DBSCAN Clutering

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python

X, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
```

```python
X.shape
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbcan=DBSCAN(eps=0.1)
```

```python
dbcan.fit_predict(X)
```

```python
dbcan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbcan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
clfLocalOutlierFactor = LocalOutlierFactor(n_neighbors=5, contamination=0.1)
y_pred_clfLocalOutlierFactor = clfLocalOutlierFactor.fit_predict(X)
```

```python
y_pred_clfLocalOutlierFactor
```

```python
clfLocalOutlierFactor.negative_outlier_factor_
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred_clfLocalOutlierFactor)
```
